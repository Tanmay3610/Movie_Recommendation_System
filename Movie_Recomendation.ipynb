```
import pandas as pd
dataset = pd.read_csv('/content/movie_dataset.csv')
```

```
dataset.head()
```

Features that we are going to make use of

```
features = ['keywords', 'cast', 'genres', 'director']
for feature in features:
  dataset[feature] = dataset[feature].fillna('')

def combine_features(row):
  return row['keywords'] + ' ' + row['cast'] + ' ' + row['genres'] + ' ' + row['director']
```

```
dataset['combined'] = dataset.apply(combine_features, axis = 1)
```

```
dataset['combined'].head()
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv = CountVectorizer()
cv_fit=cv.fit_transform(dataset['combined'])

print(cv.get_feature_names())
print(cv_fit.toarray())
```

```
cos_sim = cosine_similarity(cv_fit)
```

```
movie_user_like = 'Avatar'
indexMovie = dataset[dataset.title == movie_user_like]['index'].values[0]
```

```
similar_movies = list(enumerate(cos_sim[indexMovie]))
print(similar_movies)
```

```
recommended = sorted(similar_movies, key = lambda x:x[1], reverse=True)
```

```
number_of_recommendations = 50
for i in range(0, number_of_recommendations):
  print(dataset[dataset.index == recommended[i][0]]['title'].values[0])
```
